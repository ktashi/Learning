```python
import pandas as pd
df1 = pd.read_csv(r'C:\Users\amritt\Google Interview Practice\Data.csv')
df2 = pd.read_csv(r'C:\Users\amritt\Google Interview Practice\Data1.csv')
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insertEnd(self,newNode):
        if self.head is None:
            self.head = newNode
        else:
            lastNode = self.head
            while True:
                if lastNode.next is None:
                    break
                lastNode=lastNode.next
            lastNode.next = newNode
            
    def printList(self):
        if self.head is None:
            print("List is Empty")
            return
        currentNode = self.head
        while True:
            if currentNode is None:
                break
            print(currentNode.data)
            currentNode = currentNode.next

def mergeList (list1, list2, mergedList):
    currFirst = list1.head
    currSecond = list2.head
    while True:
        if currFirst is None:
            mergedList.insertEnd(currSecond)
            break
        if currSecond is None:
            mergedList.insertEnd(currFirst)   
            break
        if currFirst.data < currSecond.data:
            currFirstNext = currFirst.next
            currFirst.next = None
            mergedList.insertEnd(currFirst)
            currFirst = currFirstNext
        else:
            currSecondNext = currSecond.next
            currSecond.next = None
            mergedList.insertEnd(currSecond)
            currSecond = currSecondNext
    
linkedList1 = LinkedList()
for i in df1['Number']:
    iNode = Node(i)
    linkedList1.insertEnd(iNode)
print ("First List below:")
linkedList1.printList()

linkedList2 = LinkedList()
for i in df2['Number']:
    iNode = Node(i)
    linkedList2.insertEnd(iNode)
print ("Second List below:")
linkedList2.printList()
mergedList = LinkedList()
mergeList(linkedList1,linkedList2,mergedList)
print ("Merge List below:")
mergedList.printList()
```
